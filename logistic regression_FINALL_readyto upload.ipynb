```python
#this function converts the unit vector into matrix which contains only zeros and ones.....which
# this is required in order to apply multiclass classification or one vs all method one

def vector_to_matrix_of_zeros_and_ones(vect, classes): 
    m=vect.size
    Y=np.zeros((m,classes))
    flag=0
    for i in vect:
        Y[flag][i-1]=1
        flag+=1
    
    return Y   


# returns the sigmoid of the input
def sigmoid (input):
    return 1/(1+np.exp(-input))
    

#reverses the first function and compresses it into a vector
def reverse_vector_to_matrix_of_zeros_and_ones(matrix, classes):
    
    i=0
    m,k= matrix.shape
    Y=np.zeros((m,1))
    for flag in range(m):
        for i in range(classes):
            if(matrix[flag][i]==1):
                 Y[flag]=(i+1)
                 break
               
    
    return Y


#returns the cost 
def cost(hypothesis, Y):
    m=Y.shape[0]
    j= -np.sum((Y*np.log(hypothesis)+(1-Y)*np.log(1-hypothesis)))
    return j


def Z(theta,X):
    return X@theta


#inputs the data and does the changes as required to get X ,y
def input_data(s):
    print("uploading a  csv file that is appropriate for linear regression.......")

    X=pd.read_csv(s,header=None)   #load the training data

    # n=int(input("enter the number of features....."))

    X=X.to_numpy()    #converting to numpy
    i=len(X[0])       #finding the number of features
    m=X[:,0].size   
    y=X[:,0].reshape((m,1))
    X=np.delete(X,0,axis=1) 
    n=i                     #n is number of features
    u=np.ones((m,1))            #vector of ones then p.
    X=np.concatenate((u,X),axis=1)
    return X,y
```
