**DEFINING FEW FUNCTIONS**

```python
'''this function inputs the data and sorts out the prediction values and the features'''


def input_data(s):
    print("uploading a  csv file that is appropriate for KNN.......")

    X=pd.read_csv(s,header=None)   #load the training data

    # n=int(input("enter the number of features....."))

    X=X.to_numpy()    #converting to numpy
    i=len(X[0])       #finding the number of features
    m=X[:,0].size   
    y=X[:,0].reshape((m,1))
    X=np.delete(X,0,axis=1) 
    n=i                     #n is number of features
#     mean=np.mean(X,axis=1).reshape((m,1))
#     std=np.std(X,axis=1).reshape((m,1))
#     X=(X-mean)/std
#     u=np.ones((m,1))            #vector of ones then p.
#     mean=np.mean(X,axis=1).reshape((m,1))
#     std=np.std(X,axis=1).reshape((m,1))
#     X=(X-mean)/std
#     X=np.concatenate((u,X),axis=1)
#     print(X)
    return X,y




''' This is the function which returns the prediction value of the test data by finding the 
distance and applying the k nearest neighbour concept'''


def KNN(X_train , X_test) :         # root of x2-x1 square y2-y1 square from co ordinate geometry
        m_train,n_train= X_train.shape
        m_test,n_test= X_test.shape
        k=5
        y_pred=np.zeros((m_test,1))
        for i in range (m_test):
            diff=X_train-(X_test[i,:]).reshape((1,n_test))
            diff_sqr=diff**2
            sum_diff_sqr=(np.sum(diff_sqr,axis=1))
            dist=np.sqrt(sum_diff_sqr)
            argsort=np.argsort(dist)
            argsort=argsort[:5]
            count=np.zeros((n_train,1))
            for j in argsort:
                count[y_train[j]-1]+=1
            pred=(np.argmax(count))+1
            y_pred[i]=pred
            print(y_pred[i])
            
        return y_pred
    
    
    
    
''' This function is used to find the accuracy of the prediction value and the actual value
in percentage'''



def accuracy(h,y):
    m=y.size
    count=0
    for i in range(m):
        if h[i]==y[i]:
            count+=1
    accu=(count/m)*100
    return accu
```

INPUT OF THE DATA

```python
X_train , y_train= input_data('emnist-letters-train.csv')
X_test, y_test= input_data('emnist-letters-test.csv')
```

CALLING THE KNN FUNCTION 

```python
y_pred=KNN(X_train, X_test)
```

```python
accuracy(y_pred,y_test)
```
